## Benötige Informationen:

### if

```
if <Bedingung>:
    <Verarbeitungsblock>
```
Bedingung ist ein logischer Ausdruck, der entweder WAHR oder FALSCH ist.  

  
|Operator|Bedeutung      |Beispiel   |Ergebnis      |
|--------|---------------|-----------|--------------|
|<       |kleiner als    | 5 < 10    |Wahr - True   |
|        |               | 5 < 5     |Falsch - False|
|>       |größer als     | 10 > 5    |Wahr - True   |
|<=      |kleiner gleich | 5 <= 6    |Wahr - True   |
|        |               | 5 <= 5    |Wahr - True   |
|>=      |größer gleich  | 5 >= 6    |Falsch - False|
|==      |gleich         | 1 == 1.0  |Wahr - True   |
|!=      |ungleich       | 2 != 3    |Wahr - True   |
|is      |identisch      | 2 is 2    |Wahr - True   |
|is not  |nicht identisch| 4 is not 8|Wahr - True   |

  
-
  
### if - else

```
if <Bedingung>:
    <Verarbeitungsblock>
else:
    <Verarbeitungsblock>
```

### if - elif - else

```
if <Bedingung>:
    <Verarbeitungsblock>
elif <Bedingung>:
    <Verarbeitungsblock>
else:
    <Verarbeitungsblock>
```
